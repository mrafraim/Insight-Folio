<h2><a id="data-cleaning">Data Cleaning</a></h2>

#### Handing Missing Values

```python
# Check for missing values
missing_values = netflix_data.isnull().sum()
missing_values
```

```python
sns.heatmap(netflix_data.isnull())
```

```python
# 1. Handling Missing Values in 'Director' Column
# Fill missing director values with "Unknown" to indicate the absence of director information.

# Check the number of missing values before filling
missing_director_before = netflix_data['Director'].isnull().sum()

# Fill missing director values with "Unknown"
netflix_data['Director'].fillna('Unknown', inplace=True)

# Check the number of missing values after filling
missing_director_after = netflix_data['Director'].isnull().sum()

# Display the number of missing values before and after filling
print(f"Missing director values before filling: {missing_director_before}")
print(f"Missing director values after filling: {missing_director_after}")
```

```python
# 2. Handling Missing Values in 'Cast' Column
# Fill missing cast values with "Unknown" to indicate the absence of cast information.

# Check the number of missing values before filling
missing_cast_before = netflix_data['Cast'].isnull().sum()

# Fill missing cast values with "Unknown"
netflix_data['Cast'].fillna('Unknown', inplace=True)

# Check the number of missing values after filling
missing_cast_after = netflix_data['Cast'].isnull().sum()

# Display the number of missing values before and after filling
print(f"Missing cast values before filling: {missing_cast_before}")
print(f"Missing cast values after filling: {missing_cast_after}")
```

```python
# 3. Handling Missing Values in 'Country' Column
# Fill missing country values with "Unknown" to indicate the absence of country information.

# Check the number of missing values before filling
missing_country_before = netflix_data['Country'].isnull().sum()

# Fill missing country values with "Unknown"
netflix_data['Country'].fillna('Unknown', inplace=True)

# Check the number of missing values after filling
missing_country_after = netflix_data['Country'].isnull().sum()

# Display the number of missing values before and after filling
print(f"Missing country values before filling: {missing_country_before}")
print(f"Missing country values after filling: {missing_country_after}")
```

```python
# 4. Handling Missing Values in 'Release_Date' Column (Alternative Method)
# Rather than filling missing release date values with a default date, we can consider dropping rows with missing release dates if they are not critical to our analysis.

# Check the number of missing values before dropping
missing_release_date_before = netflix_data['Release_Date'].isnull().sum()

# Drop rows with missing release dates
netflix_data.dropna(subset=['Release_Date'], inplace=True)

# Reset the index after dropping rows
netflix_data.reset_index(drop=True, inplace=True)

# Check the number of missing values after dropping
missing_release_date_after = netflix_data['Release_Date'].isnull().sum()

# Display the number of missing values before and after dropping
print(f"Missing release date values before dropping: {missing_release_date_before}")
print(f"Missing release date values after dropping: {missing_release_date_after}")
```

```python
# 5. Handling Missing Values in 'Rating' Column
# Fill missing rating values with the mode (most frequent rating) of the 'Rating' column.

# Check the number of missing values before filling
missing_rating_before = netflix_data['Rating'].isnull().sum()

# Fill missing rating values with the mode
netflix_data['Rating'].fillna(netflix_data['Rating'].mode()[0], inplace=True)

# Check the number of missing values after filling
missing_rating_after = netflix_data['Rating'].isnull().sum()

# Display the number of missing values before and after filling
print(f"Missing rating values before filling: {missing_rating_before}")
print(f"Missing rating values after filling: {missing_rating_after}")
```

#### <div class='alert alert-info'> Analysis 6: Content Duration </div>

```python
 netflix_data.head()
```

```python
# Create a DataFrame that categorizes content by duration type (Minutes or Seasons)
duration_categories = netflix_data[['Category', 'Duration']]

# Separate entries with durations in minutes and seasons
minutes_category = duration_categories[duration_categories['Duration'].str.contains('min')]
seasons_category = duration_categories[duration_categories['Duration'].str.contains('Season')]

# Count the frequency of each category for both duration types
minutes_category_counts = minutes_category['Category'].value_counts().reset_index()
minutes_category_counts.columns = ['Category', 'Minutes Count']

seasons_category_counts = seasons_category['Category'].value_counts().reset_index()
seasons_category_counts.columns = ['Category', 'Seasons Count']

# Merge the two DataFrames to create a summary table
duration_summary_table = pd.merge(minutes_category_counts, seasons_category_counts, on='Category', how='outer')
duration_summary_table.fillna(0, inplace=True)

# Display the summary table
duration_summary_table
```

```python
# Table for Movie Durations in Minutes:
# Filter the dataset to include only movies (duration in minutes)
movies_data = netflix_data[netflix_data['Category'] == 'Movie']

# Extract the numeric part of the 'Duration' column for movies
movies_data['Duration'] = movies_data['Duration'].str.extract('(\d+)').astype(float)

# Calculate histogram data for movie durations, including bin edges
movie_duration_histogram_data, bin_edges = np.histogram(movies_data['Duration'], bins=30)

# Create a DataFrame to summarize movie duration data, including bin edges
movie_duration_summary_table = pd.DataFrame({
    'Bin Edges': list(zip(bin_edges[:-1], bin_edges[1:])),  # Combine bin edges into pairs
    'Movie Duration (Minutes)': movie_duration_histogram_data
})

# Display the movie duration summary table
movie_duration_summary_table
```

```python
# Analysis 6 (Movies): Content Duration in Minutes (Histogram)

# Examine the distribution of movie durations in minutes on Netflix.
# We'll create a histogram to visualize the distribution of movie durations.

# Filter the dataset to include only movies (duration in minutes)
movies_data = netflix_data[netflix_data['Category'] == 'Movie']

# Extract the numeric part of the 'Duration' column
movies_data['Duration'] = movies_data['Duration'].str.extract('(\d+)').astype(float)

# Create a histogram to visualize the distribution of movie durations in minutes
plt.figure(figsize=(12, 6))
sns.histplot(movies_data['Duration'],
             bins=30, 
             color='skyblue',
             kde=True)
plt.title("Distribution of Movie Durations in Minutes on Netflix", fontsize=16, fontweight='bold')
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")

plt.show()
```

```python
from wordcloud import WordCloud

# Combine all descriptions into a single string
all_descriptions = " ".join(netflix_data["Description"].dropna())

# Generate the word cloud
wordcloud = WordCloud(
    width=800, 
    height=400, 
    background_color='black',  # Set the background color to black
    colormap='viridis',        # Choose a color map
    max_words=2000,             # Limit the number of words displayed
    contour_width=1,           # Add contour lines
    contour_color='white',     # Set contour line color
    relative_scaling=0.5,      # Adjust word size scaling
).generate(all_descriptions)

# Display the word cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Netflix Description Word Cloud")
plt.show()
```
